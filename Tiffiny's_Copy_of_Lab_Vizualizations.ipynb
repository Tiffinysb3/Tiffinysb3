## **Data Preparation**
The first step in any data science tasks is data preparation, which is mostly about getting rid of empty/null data, outliers and general nonsensical data. This is often called 'data cleaning'. The iris data is cleaned in the following code

```python
# We will need data in order to make graphs! We will use pandas
import pandas as pd

# matplotlib is an essential whenever we are making graphs!
# Seaborn is simply a shortcut for using matplotlib!
import matplotlib.pyplot as plt

# Import Seaborn!
import seaborn as sns
```

```python
# Next, we import our dataframe using pandas. ()
iris_orig = pd.read_excel('/content/iris.xlsx')
iris_orig
```

```python
# we can drop a series of data (a row or a column)
# they're indicated by numeric value, row~0, col~1, but let's use readable names instead:
ROW = 0
COLUMN = 1

iris1 = iris_orig.drop('adapted from https://en.wikipedia.org/wiki/Iris_flower_data_set', axis=COLUMN)
iris1

# iris1 is a new dataframe, without that unwanted column (really just one element taking up a whole column...)
```

```python
# Those last two rows (index = 141 and index =142) look suspicious...Remove them!
iris = iris1.drop(axis=ROW,index=[141,142])

# Note: We are fortunate the rows with the 'bad' data was at the 'end' of the dataframe. So their deletion does not interupt the index order.
# What would have happened if these rows were at the middle or beginning of the dataframe? Well deleting them would remove the index number
# of those rows...which would result in the need for the dataframes indices to be re-numbered...rather, "reset".

# So, whenever you delete rows from a dataframe, reset the index of that dataframe
# eg iris = iris.reset_index(drop = True)
```

## **1. Scatter Plots:**
### Use the seaborn library to create scatter plots with the Iris dataset. Pick any two features (like sepal length and petal width) and plot them against each other. Explain how the data points are spread out or clustered.

```python
# Code your Scatterplot here, after which you give a short summary of how the data is spread out or clustered

import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a scatter plot between sepal_length and petal_width
sns.scatterplot(data=iris, x='sepal_length', y='petal_width', hue='species', palette='deep')
plt.title('Scatter Plot of Sepal Length vs Petal Width')
plt.show()

# Summary
"""
The scatter plot shows that the data points form distinct clusters for each species.
- Setosa points are tightly grouped with low sepal length and low petal width.
- Versicolor points form a moderate cluster in the middle range.
- Virginica points are spread at higher sepal lengths and petal widths.
This indicates clear separability between species based on these two features.
"""





```

## **2. Histograms:**
### Generate histograms for any feature of the Iris dataset (for example, petal length). Describe the distribution of this feature, noting if it's skewed, normal, or has any outliers.


```python
# Code your Histogram here, after which you give a short summary of how on the distribution of your chosen feature

# Create a histogram for petal_length
sns.histplot(data=iris, x='petal_length', bins=20, kde=True, color='skyblue')
plt.title('Histogram of Petal Length')
plt.show()

# Summary
"""
The histogram of petal length shows a multimodal distribution.
There are three distinct peaks, each corresponding roughly to one of the Iris species.
The data is not normally distributed; instead, it shows clear separation between smaller (Setosa),
medium (Versicolor), and larger (Virginica) petal lengths. There are no significant outliers.
"""




```

## **3. Box Plots:**
### Create box plots for one feature in the Iris dataset. Box Plots plots help identify the central tendency, variability, and outliers. Explain what the box plot reveals about your chosen feature.

```python
# Code your Box Plot code here, after which you give a short summary of what
# your Box plot reaveals about of your chosen feature Create a box plot for
# petal_width by species
sns.boxplot(data=iris, x='species', y='petal_width', palette='pastel')
plt.title('Box Plot of Petal Width by Species')
plt.show()

# Summary
"""
The box plot shows that:
- Setosa has the smallest petal width values with very little variation.
- Versicolor and Virginica have progressively larger petal widths, with Virginica having the highest median.
There are no major outliers, and each species has a distinct range of values, showing good separation between classes.
"""





```

## **4. Correlation Matrix:**
### Use seaborn to draw a correlation matrix of the Iris dataset's features. This matrix helps to see if any pairs of features are correlated. Discuss which features have a strong correlation or no correlation at all.

```python
# Code your Correlation Matrix here, after which you give a short summary of
# what features have positive, negative or no correlation Create a correlation
# matrix
corr = iris.drop(columns='species').corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Iris Dataset')
plt.show()

# Summary
"""
The correlation matrix shows that:
- Petal length and petal width have a very strong positive correlation (~0.96).
- Sepal length is moderately correlated with petal length and petal width.
- Sepal width has weak or negative correlations with the other features.
This means that petal features are more related to each other and useful for distinguishing species.
"""




```
