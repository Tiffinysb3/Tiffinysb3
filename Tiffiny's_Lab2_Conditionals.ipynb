# **Introduction**

### Conditionals are <i>Code Paths</i>

In Python, all decisions are made through three keywords: <tt>**if**</tt>, <tt>**elif**</tt>, and <tt>**else**</tt>  
+ These allow you to make decisions based on the value of data
+ And those decisions can direct code to take different actions.

These control structures are called ***conditionals***

In Python, **if**, **elif**, and **else** are used for making decisions in your code based on certain conditions. Here's when and how to use each one:

# **if**

Use **if** when you want to check a condition ***for the first time***. It's the starting point for a decision. Assuming the condition after the **if** statment is <font color="purple">True</font>, the code block under **if** statment gets executed. On the other hand, assuming the condition is <font color="purple">False</font>, the code block is skipped. Example:

```python
if condition1:
    # Code to execute if condition1 is True
```

# **elif**

**elif** (short for "else if"): Use **elif** after an **if-block to check additional conditions assuming the previous conditions were** <font color="purple">False</font>. You can have multiple **elif-blocks** to check multiple conditions in sequence. **elif** only gets checked when the previous **if** or **elif** conditions were <font color="purple">False</font>. Example:

```python
if condition1:
    # Code to execute if condition1 is True
elif condition2:
    # Code to execute if condition1 is False but condition2 is True
elif condition3:
    # Code to execute if condition1 and condition2 are False but condition3 is True
```

# **else**

**else**: Use **else** at the end of an **if-elif chain** to define what should happen when **none of the previous conditions were** <font color="purple">True</font>. The **else-block** does not have a condition; it catches everything that wasn't caught by the preceding **if** or **elif-blocks**. There can only be one **else** block, and it's always the last part of the decision structure. Example:

```python
if condition1:
    # Code to execute if condition1 is True
elif condition2:
    # Code to execute if condition1 is False but condition2 is True
else:
    # Code to execute if none of the above conditions are True
```

## For our examples, we will make use our trusty "random" module

```python
import random

day_in_the_week = ["Mon","Tue","Wed"]
# day_in_the_week = ["Mon","Tue","Wed","Thur","Fri"]

day = random.choice(day_in_the_week)

print(day)
```

# **Example of 'if' conditional**

```python
if day == "Tue":
  print("Taco Tuesday")
```

# **Example of 'if - elif' conditional**

```python
if day == 'Tue':
  print("Taco Tuesday")
elif day == 'Mon':
  print("We have class today")
elif day == 'Mon':
  print("We have class today")
elif day == 'Mon':
  print("We have class today")


```

# **Example of 'if - elif- else' conditional**

```python
if day == 'Tue':
  print("Taco Tuesday")
elif day == 'Mon':
  print("We have class today")
else:
  print ("The day is none of these given days")
```
