## **What is Simulations (aka Modeling)**
### In computer science, **simulation** refers to the process of creating a **model** of a real-world or hypothetical situation or system, which can then be run on a computer to study the behavior and performance of the system under various conditions.

### This model can mimic the operation of the system over time, allowing researchers, engineers, and scientists to analyze, predict, or understand complex behaviors, processes, or phenomena without the need to experiment with the real system, which might be impractical, expensive, dangerous, or impossible.

### Simulations are used across many fields and applications, including:
+ ### **Engineering and Design:** To simulate structures, vehicles, electronics, and processes to predict performance before physical prototypes are built.
+ ### **Science and Research:** To study natural systems, weather patterns, astrophysical phenomena, or biological processes.
+ ### **Economics and Social Sciences:** To model economic systems, social networks, or population dynamics.
+ ### **Healthcare and Medicine:** For simulating surgeries, understanding the spread of diseases, or the biochemical processes within the human body.
+ ### **Entertainment and Education:** In video games, virtual reality environments, and educational software to create immersive learning experiences.

###  Simulations can range from simple models that approximate behaviors using a set of equations, to highly complex simulations that require massive computational resources, like those used for weather forecasting or simulating the human brain. Key components of a simulation in computer science include the model (the mathematical and logical representation of the system), the simulation engine (software that runs the model), and the user interface (which allows interaction with the simulation).

# **Example 1: Guessing Game**

### The function below, named **guess**, is designed to simulate a computer attempting to guess a number that you've chosen, referred to as the "**hidden**" number, within a predefined range of 0 to 99 (inclusive). The function operates by repeatedly generating random guesses until it finds the hidden number. The core functionality of the function and how it operates is detailed below:

### **Function Overview**
###**Parameter/Argument**: The function takes a single parameter, hidden, which represents the number the computer is trying to guess.

###**Initial Setup**: It initializes two variables: guess is set to a value one less than hidden to ensure it starts with an incorrect guess, and number_of_guesses is set to 0 to track the number of guesses made.

###**Guessing Loop**: The function enters a while loop, where it continuously generates random numbers between 0 and 99 (inclusive) as guesses. For each iteration of the loop, it checks if the generated guess matches the hidden number.

###**Counting Guesses**: Each time the loop iterates, it increments the number_of_guesses counter by one, tracking how many guesses have been made.

###**Result**: Once the computer successfully guesses the hidden number (i.e., guess == hidden), the loop terminates, and the function returns a formatted string message. This message includes the total number of guesses it took for the computer to find the hidden number and what that number was.

```python
#
# Example while loop: the "guessing game"
#

from random import *

def guess (hidden):
    """
        have the computer guess numbers until it gets the "hidden" value
        return the number of guesses
    """
    guess = hidden -1 # start with a wrong guess + don't count it as a guess
    number_of_guesses = 0 # start with no guesses made so far...


    while guess != hidden:
      guess = choice(range(0,100)) # 0 to 99, inclusive
      number_of_guesses = number_of_guesses + 1
    return f"Wow! The computer had to loop {number_of_guesses} times before it got the number I chose, which was {hidden}"


# test our function!
guess(90)

# Run guess function using other numbers eg guess(34), guess(6) etc. Have fun and play around with it. Make sure you understand the flow of execution.
```

# **Example 2: Rolling two dice and counting doubles**

 image.png

### The **count_doubles** function (see below), simulates rolling two six-sided dice a specified number of times (**num_rolls**) and counts how many of those rolls result in doubles, meaning both dice show the same value. This type of simulation is an example of a Monte Carlo simulation, which uses repeated random sampling to obtain numerical results. The function illustrates basic concepts in probability and statistics, as well as programming constructs like loops, conditional statements, and random number generation. Here's a breakdown of how the function works:

### **Function Overview**
###**Parameter**: num_rolls specifies how many times the dice should be rolled in the simulation.

###**Variable Initialization**: numdoubles is initialized to 0. This variable tracks the number of times both dice show the same value.

###**Looping Through Rolls**: The function uses a for loop to simulate dice rolls, iterating from 0 to num_rolls - 1. For each iteration (roll), it simulates rolling two dice.

###**Dice Rolling Simulation**:
+ ### d1 and d2 are variables that represent the outcome of rolling the first and second dice, respectively.
+ ### The outcomes are determined using the choice function from the random module, which selects a random value from the specified range (1 to 6 for a six-sided die).

###**Checking for Doubles**: Inside the loop, an if statement checks if the values of d1 and d2 are equal (i.e., a double). If they are, numdoubles is incremented by one, and a smiley face emoji ("ðŸ™‚") is assigned to the variable you to indicate a double was rolled. Otherwise, you is set to a blank space.

###**Output**: For each roll, the function prints the roll number, the values of the two dice, and either a smiley face if a double was rolled or a blank space otherwise. This provides a visual representation of the simulation process and its outcomes.

###**Return Value**: After all rolls have been simulated, the function returns a formatted string that states the total number of doubles rolled during the simulation.

```python
# Example Monte Carlo simulation: rolling two dice and counting doubles
#

from random import *


def count_doubles(num_rolls):
    """
        have the computer roll two six-sided dice, counting the # of doubles
        (same value on both dice)
        Then, return the number of doubles...
    """
    numdoubles = 0 # start with no doubles so far...

    for i in range(0,num_rolls): # roll repeatedly: i keeps track
      d1 = choice(range(1,7)) # 0 to 6, inclusive
      d2 = choice(range(1,7)) # 0 to 6, inclusive
      if d1 == d2:
        numdoubles+=1  # numdoubles = numdoubles + 1
        you = "ðŸ™‚"
      else:
        you = ""
      print("run", i, "roll:", d1, d2, you)

    return f'The number of times the dice got doubles is {numdoubles} times'


count_doubles(100)


```
