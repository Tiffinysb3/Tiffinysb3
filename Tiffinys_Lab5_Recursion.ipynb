## **Introduction**
A very powerful approach to problem-solving is _recursion_. <font color = "magenta">Recursion is a programming concept where a **function calls itself directly or indirectly** to solve a problem.</font> It's like solving a big problem by breaking it down into smaller, more manageable problems of the same type.


Recursion finds and uses a situation's _self-similarity_ to solve a problem.  Self-similarity is the property of a situation or problem that allows it to be described or solved in terms of smaller instances of itself.
+ self-similarity sometimes feels less "natural" than sequencing (i.e., solving problems through a sequence of steps or iterations)
+ however, there are problems where self-similarity is a much better fit.
+ also, if a problem is solvable, <font color = "magenta">it's _always_ possible to solve it with recursion</font>
  + to be fair, it's also always possible to solve it with sequence-based approaches (iteration)
  + that said, _humans_ who are familiar with both -- they tend to have a much bigger set of solution-techniques!

## **Examples of Recursion**


## **1.Factorial**
## The factorial of a number (n!) is the product of all positive integers less than or equal to n.
## For example, 5! = 5 * 4 * 3 * 2 * 1 = 120. In recursion, the factorial function calls itself with a smaller value each time:

```python
# factorial of 5 = 5 x 4 x 3 x 2 x 1

def factorial(n):
  # Base case:
  if n == 1:
    return 1
  else:
    return n * factorial(n-1)


# Now use the function to find the factorial of and arbitrary number n eg 3
d = factorial(3)

print(d)
```

## **2.Fibonacci Sequence**
## The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding ones, usually starting with 0 and 1. That is, F(0)=0, F(1)=1, and F(n)=F(n-1)+F(n-2) for n > 1. Here's how you can generate the Fibonacci sequence using recursion:

```python
def fibonacci(n):
    # Base cases: the first two numbers of the sequence
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        # Recursive case: sum of the two preceding numbers
        return fibonacci(n-1) + fibonacci(n-2)




# Now use the function to find the fibonacci sequence of and arbitrary number n eg 10
W = fibonacci(10)

print(W)
```

## **3.Replacing characters of strings**
## This function replaces the letter 's'in a word with the special symbol "$"

## For this recursive function, lets take a stroll down memory lane:
## Recall when we worked with **string slicing**, if we have a string **s**, the index **s[x:]** means "<font color="magenta">give me all the characters/letters in the string starting from the character with index x</font>". Therefore, any character/letter value belonging to indexes before x is ignored.
## So, for example, if s = "**[supercalifragilisticexpialidocious](https://youtu.be/uZNRzc3hWvE?si=xy9qzKV5V-bJWBff)**", then the string slice **s[1:]** will *drop* the first letter ('**s**', which is at index 0) and return the rest of the letters/characters of the string ie "**upercalifragilisticexpialidocious**"




```python

def replace_s(word):
  if word == "":
    return word
  elif word[0] == 's':
    return "$" + replace_s(word[1:])
  else:
    return word[0] + replace_s(word[1:])


x = replace_s('Pains and Joys')
y = replace_s("supercalifragilisticexpialidocious")
print(y)
print(x)
```

## **4.All about vowels**

## The next set of examples show recursion functions that perform different tasks on vowels of a string

### **a. Function that Counts number of vowels in a string**

```python
def vwl(s):
    """vwl returns the number of vowels in s
       Argument: s, which will be a string
    """
    if s == '':
        return 0   # no vowels in the empty string
    elif s[0]  in 'aeiou':
        return 1 + vwl(s[1:])   # count 1 for the vowel. Oke note: remember s[1:] means 'the rest of the elts in s after the first elt (s[0])
    else:
        return 0 + vwl(s[1:])   # The 0 + isn't necessary but looks nice

#
# Tests!
print( "vwl('sequoia') should be 5 <-> ", vwl('sequoia') )
print( "vwl('bcdfg') should be 0 <-> ", vwl('bcdfg') )
print( "vwl('e') should be 1 <-> ", vwl('e') )
print( "vwl('This sentence has nine vowels.') should be 9 <-> ", vwl('This sentence has nine vowels.') )
```

### **b.Function that returns the vowels of a string**

```python
def keepvwl(s):
    """keepvwl returns the vowels in s
       Argument: s, which will be a string
       Return value: a string with all of s's vowels, in order
    """
    if s == '':
        return ''   # return the empty string
    elif s[0] in 'aeiou':
        return s[0] + keepvwl(s[1:]) # keep s[0], since it's a vowel
    else:
        return '' + keepvwl(s[1:])   # '' isn't necessary but fits conceptually!


# Tests
print( "keepvwl('sequoia') returns <-> ", keepvwl('sequoia') )
print( "keepvwl('bcdfg') returns <-> ", keepvwl('bcdfg') )
print( "keepvwl('e') returns <-> ", keepvwl('e') )
print( "keepvwl('This sentence has nine vowels.') returns <-> ", keepvwl('This sentence has nine vowels.') )
```

### **c. Function that drops all the vowels in a string**

```python
def dropvwl(s):
    """dropvwl returns the non-vowels in s.  Note that "non-vowels" includes
         characters that are not alphabetic.
       Argument: s, which will be a string
       Return value: s with the vowels removed
    """
    if s == '':
        return ''   # return the empty string
    elif s[0] in 'aeiou':
        return '' + dropvwl(s[1:])   # drop s[0], since it's a vowel
    else:
        return s[0] + dropvwl(s[1:])   # keep s[0], since it's NOT a vowel

# Tests
print( "dropvwl('sequoia') returns <-> ", dropvwl('sequoia') )
print( "dropvwl('bcdfg') returns <-> ", dropvwl('bcdfg') )
print( "dropvwl('e') returns <-> ", dropvwl('e') )
print( "dropvwl('This sentence has nine vowels.') returns <-> ", dropvwl('This sentence has nine vowels.') )
```
